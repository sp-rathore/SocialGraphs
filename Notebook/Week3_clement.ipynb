```python
import numpy as np

prob = [0,0.01,0.03,0.05,0.1,0.2]
d = [[x, []] for x in prob]
l = []

for k in d:
    for i in range(50):
        gg = nx.watts_strogatz_graph(500, 4, k[0])
        k[1] += [nx.average_shortest_path_length(gg)]
        print k[1]
    l += [[k[0], np.mean(k[1]), np.std(k[1])]]

print l
```
