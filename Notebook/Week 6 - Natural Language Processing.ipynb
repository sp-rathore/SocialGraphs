#### What is a bigram? How does it relate to collocations. Explain in your own words.

Bigram refers to word pairs that could be most commonly used with the other word. e.g. 'more' 'than', 'too' etc..
Collocation refers to sequence of words that occur together often. Thus more than is a collocation, whereas the too than is not.
